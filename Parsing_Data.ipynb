```python
with open("Initial_data.txt", encoding = "utf-8") as f:
    data = f.read()
```

```python
data
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
```

```python
chunks[2]
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split("\n")
    username = sep_chunk[0]
    no_of_posts = sep_chunk[1]
    no_of_followers = sep_chunk[2]
    no_of_following = sep_chunk[3]
    name = sep_chunk[4]
    if(len(sep_chunk)> 5):
        type_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
            type_of_page = "Unknown"
            bio = ""
    # print(username, no_of_followers, no_of_following, no_of_posts, name, type_of_page, bio, sep= "\n")

    return {"username": username, "no_of_followers": no_of_followers, "no_of_following": no_of_following, "no_of_post": no_of_posts, 
            "name": name, "type_of_page": type_of_page, "bio": bio}
print(parse_chunk(chunks[0]))
```

```python
all_chunks = [] 
for chunk in chunks:  
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks, indent = 4)
print(s)
```
